```python
%matplotlib inline
import scipy 
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.loadtxt('/home/ubuntu/Notebooks/datasets/ExtramaritalAffairsData.csv', delimiter=",", skiprows=2)
```

```python
X = np.zeros(A.shape) #Matrix full of zeroes with the same dimension as matrix A.
X = X[:,0:8] #Then we take all rows, and skip column 9.
```

```python
for i in range (8):
    X[:,i] = (A[:,i]-np.min(A[:,i])/(np.max(A[:,i]) - np.min(A[:,i])))
#Max-Min normalization by columns. Each column has a feature (such as age, and so on...)
```

```python
X = X[0:100,:] #First 100 individuals.
Y = A[0:100,-1] #Last column from first 100 individuals. (Infidelity row)
```

```python
#Histograma
plt.figure(figsize=(15,5))
plt.title('Relaciones Extramaritales')
plt.xlabel(u'Infidelidades por año')
plt.ylabel(u'Número de individuos')
h = plt.hist(Y,100,color="pink") #number of bins
```

```python
#Gráfico de dispersión
plt.figure(figsize=(15,5))
plt.title('Relaciones Extramaritales')
plt.xlabel(u'Edad')
plt.ylabel(u'Infidelidades por año')
plt.scatter(A[:,1],A[:,-1],s=150,color="pink") #s=size of points
```

```python
#Gráfico de dispersión
plt.figure(figsize=(15,5))
plt.title('Relaciones Extramaritales')
plt.xlabel(u'Nivel de educación')
plt.ylabel(u'Infidelidades por año')
plt.scatter(A[:,5],A[:,-1],s=150,color="pink") #Error en el PDF
```

```python
#Dendograma (Diagrama de árbol)
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = linkage(X,'ward') #Performs hierarchical/agglomerative clustering. 
#It uses the Ward variance minimization algorithm. (Minimum distances)
```

```python
plt.figure(figsize=(15,5))
plt.title(u'Diagrama de árbol jerarquico')
plt.xlabel(u'Individuos')
plt.ylabel(u'Distancia')
dendrogram(Z,leaf_rotation="90",leaf_font_size=7) #leaf_rot: x-labels 
plt.show()
```
